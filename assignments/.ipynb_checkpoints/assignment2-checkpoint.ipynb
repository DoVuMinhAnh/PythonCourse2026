## Task 2

### Part a

```python
import requests

tickers = ["META", "AAPL", "AMZN", "NFLX", "GOOG"]
all_data = []
for t in tickers: 
    url = (
        f"https://api.marketstack.com/v2/eod"
        f"?access_key={api_key}"
        f"&symbols={t}"
        f"&date_from=2026-03-01"
        f"&date_to=2026-03-31"
        f"&limit=50")
    response = requests.get(url)
    data = response.json()
    all_data.extend(data["data"])
print(all_data[:5])
```

### Part b

```python
data = pd.DataFrame(all_data)

data = data.sort_values("date").reset_index()

print(data.head())
```

### Part c

```python
#summary
summary = data.groupby("symbol")["close"].describe()
print (summary)
```

```python
#return percentage
first = data.groupby("symbol")["close"].first()
last = data.groupby("symbol")["close"].last()

returns = ((last - first) / first) * 100
returns = returns.reset_index()
returns.columns = ["symbol", "return_%"]
returns = returns.sort_values("return_%", ascending=False)

print(returns)
print(f"\nBest performer:  {returns.iloc[0]['symbol']} ({returns.iloc[0]['return_%']}%)")
print(f"Worst performer: {returns.iloc[-1]['symbol']} ({returns.iloc[-1]['return_%']}%)")
```

## Task 3

### Part b

```python
sns.barplot(data = returns, x = "return_%" , y = "symbol")

plt.title("Overall return percentage for each stocks in March 2026")
plt.xlabel("Return percentage")
```

### Part c

```python
pivot = data.pivot(index="symbol", columns="date", values="volume")

pivot.columns = pd.to_datetime(pivot.columns).strftime("%d")

sns.heatmap(data = pivot, cmap="YlOrRd",
            linewidths=0.5, cbar_kws={"label": "Volume"})
```

### Part c

```python
queryc = """
    SELECT c.CompanyName,
           o.OrderID,
           o.OrderDate,
           o.ShipCountry,
           od.UnitPrice * od.Quantity AS LineTotal
    FROM Orders o
    JOIN Customers c ON o.CustomerID = c.CustomerID
    JOIN 'Order Details' od ON o.OrderID = od.OrderID
    WHERE o.ShipCountry = 'Germany'
    """

germany =  pd.read_sql_query(queryc, con)
print(germany.head())
```

## Task 6

```python
orders_per_customer = (germany.groupby("CompanyName")["OrderID"]
                                 .nunique()
                                 .reset_index(name="OrderCount")
                                 .sort_values("OrderCount", ascending=False)
                                 .head(10))

sns.barplot(data = orders_per_customer, x = "OrderCount", y = "CompanyName")
plt.title("Top 10 customers by number of orders shipped to Germany")
```

### Part b

```python
germany["OrderDate"] = pd.to_datetime(germany["OrderDate"], format = "mixed")
daily_sales = germany.groupby("OrderDate")["LineTotal"].sum().reset_index()
print(daily_sales.head())

sns.kdeplot(data=daily_sales, x="LineTotal", fill=True)

plt.title("Distribution of daily total sales to Germany")
plt.xlabel("Daily sales (USD)")
plt.ylabel("Density")
```

### Part c

```python
query6c = """
    SELECT cat.CategoryName,
           SUM(od.UnitPrice * od.Quantity) AS TotalRevenue
    FROM 'Order Details' od
    JOIN Products p ON od.ProductID = p.ProductID
    JOIN Categories cat ON p.CategoryID = cat.CategoryID
    GROUP BY cat.CategoryName
"""

ger_revenue = pd.read_sql_query(query6c, con)

sns.barplot(data=ger_revenue, x="TotalRevenue", y="CategoryName")
plt.title("Total revenue by product category")
```

### Part c

```python
import time

##SQL
startsql = time.time()

query_sql = """
    SELECT cat.CategoryName,
           SUM(od.UnitPrice * od.Quantity) AS TotalRevenue
    FROM 'Order Details' od
    JOIN Products p ON od.ProductID = p.ProductID
    JOIN Categories cat ON p.CategoryID = cat.CategoryID
    GROUP BY cat.CategoryName
    ORDER BY TotalRevenue DESC
"""

resultdf = pd.read_sql_query(query_sql, con)
endsql = time.time()
sql_time = endsql - startsql
print(f"SQL time: {sql_time:.4f} seconds")

##Pandas
startpandas = time.time()

order_details = pd.read_sql_query("SELECT * FROM 'Order Details'", con)
merged7c = (pd.merge(pd.merge(order_details, productdf, on="ProductID"), 
                             categoriesdf, on="CategoryID")
            .assign(Revenue=lambda x: x["UnitPrice_x"] * x["Quantity"])
            .groupby("CategoryName")["Revenue"]
            .sum()
            .reset_index())

endpandas = time.time()

pandas_time = endpandas - startpandas
print(f"Pandas time: {pandas_time:.4f} seconds")

if sql_time < pandas_time:
    print("SQL is faster")
else:
    print("Pandas is faster")
```

## Task 8

```python
##wide dataset 
wide_data = pd.DataFrame({
    "store": ["Helsinki", "Tampere", "Turku"],
    "Q1_sales": [12000, 8500, 9200],
    "Q2_sales": [13500, 9100, 9800],
    "Q3_sales": [11800, 8800, 10100],
    "Q4_sales": [15000, 10200, 11500]
})

##long dataset
long_data = pd.DataFrame({
    "month": ["Jan", "Jan", "Feb", "Feb", "Mar", "Mar"],
    "product": ["Laptop", "Phone", "Laptop", "Phone", "Laptop", "Phone"],
    "units_sold": [45, 120, 50, 110, 55, 130]
})
```

### Part a

```python
print("Before conversion")
print(wide_data)
print()

wide_to_long = pd.melt(wide_data,
                      id_vars="store", 
                        var_name="quarter", 
                        value_name="sales")
print("After conversion")
print(wide_to_long)
```

### Part b

```python
print("Before conversion")
print(long_data)
print()

long_to_wide = long_data.pivot(columns = "product",
                               values = "units_sold",
                               index = "month").reset_index()
print("After conversion")
print(long_to_wide)
```

### Part c

```python
#back to wide
print("Converted long version:")
print(wide_to_long)
print()

wide_recover = wide_to_long.pivot(columns = "quarter",
                                  index = "store",
                                  values = "sales")
print("Recovered wide:")
print(wide_recover)
```

```python
#back to long
print("Converted wide version:")
print(long_to_wide)
print()

long_recover = pd.melt(long_to_wide, 
                    id_vars = "month",
                    var_name = "product", 
                    value_name = "units_sold")
print("Recovered long:")
print(long_recover)
```
